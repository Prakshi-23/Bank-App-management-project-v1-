```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('HDFC BANK CUSTOMER INFO.xlsx')
```

```python
df
```

```python
df.rename(columns={'target balance':'target_balance'},inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['monthly_income']=df['monthly_income'].fillna('None')
```

```python
df.columns
```

```python
df=df[['gender','age', 'occupation', 'monthly_income', 'marital_status', 'education_level','target_balance']]
```

```python
df.rename(columns={'target_balance':'balance'},inplace=True)
```

```python
cat_col=df.select_dtypes(include='object').columns.tolist()
num_col=df.select_dtypes(include='number').columns.tolist()
```

```python
for i in cat_col:
    print(i)
    plt.figure(figsize=(10,6))
    sns.countplot(data=df,x=i)
    plt.show()
```

```python
for i in num_col:
    print(i)
    plt.figure(figsize=(18,6))
    sns.scatterplot(data=df,x=i,y='balance')
    plt.show()
```

```python
df=pd.get_dummies(df,columns=cat_col,dtype=int,drop_first=True)
```

```python
df
```

```python
X=df.drop('balance_>50K',axis=1)
y=df['balance_>50K']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=55,stratify=y)
```

```python
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier
```

```python
# pip install xgboost
```

```python
from xgboost import XGBClassifier
```

```python
ada=AdaBoostClassifier()
```

```python
ada.fit(X_train,y_train)
```

```python
ada_pred=ada.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,ada_pred))
```

```python
gb=GradientBoostingClassifier()
```

```python
gb.fit(X_train,y_train)
```

```python
gb_pred=gb.predict(X_test)
```

```python
print(classification_report(y_test,gb_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
log_pred=lr.predict(X_test)
```

```python
print(classification_report(y_test,log_pred))
```
